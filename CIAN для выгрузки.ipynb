***Выгрузка данных***


Теперь выгрузим все характеристики и их значения в csv файл, проходя по ссылки каждого объявления и вытаскивая, нужные нам данные. По идее в ранге должны стоять данные (0, 1508), но комп решил выключиться, и выгрузка данных остановилась на 1505 объявлении, поэтому в данном коде мы берем range(1506,1508) (до этого range был 1508).

После изучения структуры объявлений на сайте Циана, нам стало понятно, какие данные / блоки данных, нужно выгружать. 

Также мы столкнулись с трудностью, что не во всех объявлениях есть все характеристики, поэтому код ломался на 85 объявлении. Для того, чтобы он прошел и выгрузил все данные вне зависимости есть ли та или иная характеристика, мы используем команду except.

```python
import csv
browser = webdriver.Chrome(r'C:/Users/sashka/Documents/Универ/Андан')
#выгрузка и запись на 10 первых
with open('last_try.csv', 'a', newline='\n', encoding="utf-8") as data:
    writer = csv.writer (data)
    for link in tqdm(range(1506,1508)):
        para_to=[]
        a=lines[link][:-1]
    #создаём первуую колонку (номер объявления)
        para_to.append(a.split('/')[-2])
        browser.get(a)
        sleep(2)
    #зашли на страницу объявления
        try:
            para_to.append((browser.find_element(By.CLASS_NAME, "a10a3f92e9--title--vlZwT")).text[0])
        except:
            para_to.append('none')
    #количество комнат
        try:
            para_to.append((browser.find_element(By.CLASS_NAME, "a10a3f92e9--price--Pg6fn")).text)
        except:
            para_to.append('none')
    #цена
        try:
            para_to.append((browser.find_element(By.CLASS_NAME, "a10a3f92e9--underground_link--VnUVj")).text)
        except:
            para_to.append('none')
        try:
            para_to.append((browser.find_element(By.CLASS_NAME, "a10a3f92e9--underground_time--YvrcI")).text)
        except:
            para_to.append('none')
    #метро+время до метро, берём первое, потому что циан сам сортирует время по возрастанию (нужно добавить разделение на машине/пешком)
        try:
            c=(browser.find_element(By.CLASS_NAME, "a10a3f92e9--container--tqDAE")).text
            para_to.append(c.split('\n'))
        except:
            para_to.append('none')
            
        try:
            d=(browser.find_element(By.CLASS_NAME, "a10a3f92e9--container--rGqFe")).text
            para_to.append(d.split('\n'))
        except:
            para_to.append('none')
            
        writer.writerow(para_to)
        
data.close()
```
