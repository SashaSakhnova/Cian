Как мы видим, характеристики были выгружены блоками и заключены все в двух столбцах, дальнейшими действиями мы будем вычленять каждое из наблюдений и выносить в отдельные столбцы. Используем для этого словари, в которых ключами будет сами наблюдения

```python
def func1(x):
    res = x.strip('][').split(', ')
    #zip - сжимаем данные
    d = dict(zip(res[::2], res[1::2]))
    return d
```

```python
def func2(x):
    res = x.strip('][').split(', ')
    res.pop(0)
    d = dict(zip(res[::2], res[1::2]))
    return d
```

```python
df['list1'] = df['list1'].apply(lambda x: func1(x))
```

```python
df['list2'] = df['list2'].apply(lambda x: func2(x))
```

Теперь обработаем наблюдения, которые содержат в себе информацию о метраже. В начале работы мы столкнулись с ошибкой изменения типа, так как наша таблица в качестве разделительного знака содержит запятые, а для того, чтобы мы могли преобразовать данные к типу ***`float`*** нужны точки.

```python
df['WC'].value_counts()
```

```python
def replace_wc(row):
    if np.isnan(row['WC']):
        area = float(row['total area(м²)'])
        if area < 70:
            return 1
        elif area < 150:
            return 2
        elif area < 300:
            return 3
        else:
            return 4
    else:
        return row['WC']

df['WC'] = pd.to_numeric(df['WC'], errors='coerce')
df['total area(м²)'] = pd.to_numeric(df['total area(м²)'], errors='coerce')
df['WC'] = df.apply(lambda row: replace_wc(row), axis=1)
```
